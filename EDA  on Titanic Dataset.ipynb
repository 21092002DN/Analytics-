# Feature engineering
### including family size and family type

```python
data['family_size']=data['Parch']+data['SibSp']
```

```python
data
```

```python
def family_type(number):
   if number==0:
     return 'alone'
   elif number>0 and number<=4:
      return 'medium'
   else: 
      return 'large'
    
 
```

```python
 data['family_type']=data['family_size'].apply(family_type) 
```

```python
data
```

```python
sns.heatmap(data.corr())
```

```python
data.drop(columns=['Parch','SibSp'])
```

```python
# survived with family size
pd.crosstab(data['family_size'],data['Survived']).apply(lambda r: round((r/r.sum())*100,1),axis=1)
```

```python
sns.countplot(x='Survived',hue='family_size',data=data)
```
